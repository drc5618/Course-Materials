### what is your idea?

We are planning a car rental business near or within an airport. People will rent cars for between 3 days to 2 weeks.

### what makes it unique?

People can rent other vehicles such as motorbikes, four wheelers, and RVs in addition to cars. The vehicles also have really obnoxious horns.

### is this a brand-new product or business, or are you borrowing from another business?

We are borrowing from another business. Car rentals already exist, but the variety of vehicles and the vehicles' horns make it unique.

### describe a typical business interaction or a day at your business.

A person who arrives or will arrive from the airport can order online or in person. Customers will arrive at the front desk where they will tell the workers what type of motor vehicle they want. In addition, they will tell the worker how long they will be renting said motorvehicles. Then, a worker will bring out the requested motorvehicle to the customer.
